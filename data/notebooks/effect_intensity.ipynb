```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas
import numpy as np

proc_dir = "../processed/"
```

```python
data = pandas.read_csv(proc_dir + "intensities.csv", index_col=0)
```

```python
print(data)
```

```python
means = data.mean(axis=0)
stdevs = data.std(axis=0)
```

```python
print(means)
print(stdevs)
```

```python
print(means['heavy_rain'])
print(stdevs['heavy_rain'])
```

```python
print(means['light_rain'])
print(stdevs['light_rain'])
```

```python
h75, h25 = np.percentile(data['heavy_rain'], [75, 25])
heavy_iqr = h75 - h25
print(heavy_iqr)

l75, l25 = np.percentile(data['light_rain'], [75, 25])
light_iqr = l75 - l25
print(light_iqr)
```

```python

fig = plt.figure(figsize=(12, 4))


bplot = plt.boxplot(data, patch_artist=True, labels=['Light Rain', 'Wind', 
                                                     'Thunder', 'Blizzard', 'Clear Sky', 'Cloudy', 'Heavy Rain', 'Snow'],
                   medianprops=dict(color='red'))

plt.title("Perceived Intensity of Sound Effects")
plt.xlabel("Weather Condition")
plt.ylabel("Mean Intensity Rating (Out of 5)")

plt.grid()

colours = sns.color_palette('pastel', n_colors=8)
for patch, color in zip(bplot['boxes'], colours):
        patch.set_facecolor(color)
```
