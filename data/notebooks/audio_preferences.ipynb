```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas
import numpy as np

proc_dir = "../processed/"
data = pandas.read_csv(proc_dir + "sound_preferences.csv")
```

```python
print(data['most_effective'].values)
most_effective = data['most_effective'].value_counts()
print(most_effective)
plt.pie(most_effective, labels=most_effective.keys(), colors=sns.color_palette('Set2'), autopct='%1.1f%%', explode=(0.05, 0.05, 0.05, 0.05, 0.05, 0.05))
plt.title("Which sound was the most effective at \nconveying its respective weather condition?")
```

```python
least_effective = data['least_effective'].value_counts()
print(least_effective)
plt.pie(least_effective, labels=least_effective.keys(), colors=sns.color_palette("Set2"), autopct='%1.1f%%',explode=(0.05, 0.05, 0.05, 0.05, 0.05))
plt.title("Which sound was the least effective at \nconveying its respective weather condition?")
```
