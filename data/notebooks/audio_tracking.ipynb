```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas
import numpy as np

proc_dir = "../processed/"
```

```python
data = pandas.read_csv(proc_dir + "audio_tracking.csv")


frame_entries = data[data['headset'] == 'frames']
headphone_entries = data[data['headset'] == 'nc700']

print(headphone_entries)
print(frame_entries)
```

```python
print(headphone_entries.mean())
print(headphone_entries.std())

print(frame_entries.mean())
print(frame_entries.std())
```

```python
fig = plt.figure(figsize=(3,3))
plt.grid()

ax = fig.add_axes([0,0,1,1])
headsets = ['NC700 Headphones', 'Bose Frames']

means = pandas.DataFrame({'nc700': [headphone_entries.mean()[0]], 'frames': [frame_entries.mean()[0]]})
stdvs = pandas.DataFrame({'nc700': [headphone_entries.std()[0]], 'frames': [frame_entries.std()[0]]})
print(means)
print(stdvs)
ax.set_ylim(1, 5)
barplot = ax.bar(headsets, list(means.values[0]),
 color=sns.color_palette('Pastel2'), yerr=stdvs,
 capsize=10)

ax.bar_label(barplot, label_type='center')
plt.title("How easy was it to determine where in \n 3D space a sound was coming from?")
plt.xlabel("Headset")
plt.ylabel("Mean User Rating (Out of 5)")

plt.show()
```
